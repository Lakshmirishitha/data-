```python
import matplotlib.pyplot as plt
x=[1,2,3]
y=[4,5,5]
plt.plot(x,y)
plt.title("info")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
x=[10,20,30,40]
mylabels=["Apples","Bananas","cherries","dates"]
mycolors=["black","hotpink","b","r"]
plt.pie(x,labels=mylabels,colors=mycolors)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=([35,25,25,15])
mylabels=["Apples","Banana","Cherries","Dates"]
myexplode=[0.2,0,0.6,0.5]
plt.pie(x,labels=mylabels,explode=myexplode,shadow=True)
plt.show()
```
